```python
# 3. 손흥민 뉴스기사 만들기

from datetime import datetime 
import random 

sons_news = {
    'place' : 'null',
    'time' : 'null',
    'oppenent' : 'null',
    'goals' : 'null',
    'aids' : 'null',
    'score_me' : 'null',
    'score_you' : 'null'
}
sons_news_question = {
  'place' : "경기가 열린 곳은?",
  'time' : "경기가 열린 시간은?",
  'oppenent' : "상대 팀은?",
  'goals' : "손흥민은 몇골을 넣었나요?",
  'aids' : "도움은 몇개 인가요?",
  'score_me' : "손흥민 팀이 넣은 골 수는?",
  'score_you' : "상대 팀이 넣은 골 수는?"
}

int_value = list(sons_news.keys())[3:]

#경기 결과 입력 받는 곳 
for i in sons_news.keys():
  sons_news[i] = input(sons_news_question[i])
  
print("\n\n")

for i in int_value:
  while not(sons_news[i].isnumeric()) :
    print("잘못된 입력입니다.",end = '')
    sons_news[i] = input(sons_news_question[i])

news = "[프리미어 리그 속보(" + str(datetime.now()) + ")]\n"
news += "손흥민 선수는 " + sons_news['place'] +"에서 " + sons_news['time'] + "에 열린 경기에 출전하였습니다.\n"
news += "상대 팀은 "+ sons_news['oppenent'] + "입니다.\n"

if sons_news['score_me'] > sons_news['score_you'] :
  news += "손흥민 선수의 팀이 " + sons_news['score_me'] + "골을 넣어 " + sons_news['score_you'] + "골을 넣은 상대 팀을 이겼습니다.\n"
elif sons_news['score_me'] < sons_news['score_you'] :
  news += "손흥민 선수의 팀이 " + sons_news['score_me'] + "골을 넣어 " + sons_news['score_you'] + "골을 넣은 상대 팀에게 졌습니다.\n"
else :
  news += "두 팀은" + sons_news['score_me'] + "대" + sons_news['score_you'] + "로 비겼습니다.\n"

if int(sons_news['goals']) > 0 and int(sons_news['aids']) > 0 :
  news += "손흥민 선수는 " + sons_news['goals'] + "골에 도움 " + sons_news['aids'] + "개로 승리를 크게 이끌었습니다.\n"
elif int(sons_news['golas']) > 0 and int(sons_news['aids']) == 0 :
  news += "손흥민 선수는 " + sons_news['goals'] + "골로 승리를 이끌었습니다.\n"
elif int(sons_news['goals']) == 0 and int(sons_news['aids']) > 0 :
  news += "손흥민 선수는 골은 없지만 도움" + sons_news['aids'] + "개로 승리하는 데 공헌하였습니다.\n"
else:
  news += "아쉽게도 이번 경기에서 손흥민 선수의 발끝은 침묵을 지켰습니다.\n"

print(news)

```

```python
 # 연습문제 8 무작위 행보는 이동 방향을 무작위로 선택해 이동 궤적을 구한다. 이를 이용해 다음사항을 수행하시오.
 # (1) 동서남북 네 방향에 대해 동일한 확률로 이동하는 무작위 행보 프로그램을 작성하시오. 이동 궤적은 Turtle graphic 라이브러리를 사용하여 그린다.
import ColabTurtle.Turtle as t
import random

t.initializeTurtle(initial_speed=5) 
t_spd = 100
t.color(0,0,255)

CardinalPoints = {
    "동" : (t_spd, 0),
    "서" : (-t_spd , 0),
    "남" : (0 , -t_spd),
    "북" : (0 , t_spd),
    }
RandomCardinalPoints = random.choice(list(CardinalPoints.keys()))
nextPos = [x+y for x,y in zip(t.pos(), CardinalPoints[RandomCardinalPoints])]

t.goto(tuple(nextPos))
```
