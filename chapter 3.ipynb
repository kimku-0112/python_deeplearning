```python
# 연습문제 7 *은 정상품이고 -은 불량품이다. 생산 라인에 설치된 자동 검사 시스템이 12개 생산품에 대해 다음과 같이 P또는 N으로 예측했다.
# P는 불량, N은 정상이다. 이를 통해 1.~3.의 결과를 적으시오.
class Product():
    data = []
    result = []
    datasize = 0
    def __init__(self, rawdata):
        self.datasize = len(rawdata) / 2
        if len(rawdata) & 1: print("wrong data")
        else :
            for i in range(len(rawdata)):
                if i & 1: self.data.append(rawdata[i] == 'N')
                else : self.result.append(rawdata[i] == '*')


def main():
    #product_data = ['*N', '*N', '*P', '-P', '*N', '*N', '-P', '*N', '*N', '*N', '-N', '*N']
    product_data = '*N*N*P-P*N*N-P*N*N*N-N*N'
    product = Product(product_data)

    print(product.data)
    print(product.result)
    print(product.datasize)
    
    
    accuracy = 0 # 전체 샘플 중에서 맞힌 샘플의 비율 
    specificity = 0 # 부정의 샘플중에서 맞힌 샘플의 비율
    sensitivity = 0 # 긍정의 샘플중에서 맞힌 샘플의 비율
    percision = 0 # 긍정의 예측값 중에서 맞힌 샘플의 비율
    recall = 0 # # 긍정의 샘플중에서 맞힌 샘플의 비율
    
    sample_P = 0
    sample_N = 0
    result_P = 0
    result_N = 0
    
    for i in range(len(product.data)): 
        sample_P += product.data[i]
        sample_N += not product.data[i]
        result_P += product.result[i]
        result_N += not product.result[i]
        
        accuracy += (product.data[i] ==  product.result[i])
        specificity += (not product.data[i]) & (product.data[i] == product.result[i])
        sensitivity += product.data[i] & (product.data[i] == product.result[i])
        percision += product.result[i] & (product.data[i] == product.result[i])
        
    recall = sensitivity
    
    accuracy /= (product.datasize/100)
    specificity /= sample_N
    sensitivity /= sample_P
    percision /= result_P
    recall /= sample_P

    print('1. 정확률은 얼마인가?')
    print('정확률 : ', accuracy, ' %')

    print('2. 정밀도와 재현율은 얼마인가?')
    print('정밀도 : ', percision)
    print('재현율 : ',recall)

    print('3. 특이도와 민감도는 얼마인가?')
    print('특이도 : ',specificity)
    print('민감도 : ',sensitivity)

if __name__ == '__main__':
    main()
```

```python
# 연습문제 10 [프로그램 3-1]이 SVM뿐만 아니라 결정 트리 모델까지 적용하도록 확장하시오. 
# 이때 훈련 집합을 테스트 집합으로 간주해 SVM과 결정 트리의 정확률을 측정하고 비교하시오.
from sklearn import datasets
from sklearn import svm 
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd 
import numpy as np

iris = datasets.load_iris()
x_train, x_test, y_train, y_test = train_test_split(iris.data,
                                                    iris.target,
                                                    test_size=0.2,
                                                    random_state=15)

svm_model = svm.SVC(gamma = 0.1, C = 10)          # svm 분류 모델 SVC 객체 생성하고 
svm_model.fit(x_train, y_train)                   # iris 데이터로 학습 
svm_predicted = svm_model.predict(x_test)
svm_accuracy = accuracy_score(y_test,svm_predicted)

decision_tree_model = DecisionTreeClassifier(random_state=1234)
decision_tree_model.fit(x_train,y_train)
decision_tree_predicted = decision_tree_model.predict(x_test)
decision_tree_accuracy = accuracy_score(y_test,decision_tree_predicted)

print(f'svm Mean accuracy score: {svm_accuracy:.3}')
print(f'decision tree Mean accuracy score: {decision_tree_accuracy:.3}')

print("1번째 샘플과 2번째 샘플의 부류는 \t\t", np.array([iris.target_names[i] for i in y_test])[0:2])
print('svm 모델이 예측한 1번째 샘플과 2번째 샘플의 부류는',np.array([iris.target_names[i] for i in svm_predicted])[0:2])
print('decision tree 모델이 예측한 1번째 샘플과 2번째 샘플의 부류는',np.array([iris.target_names[i] for i in decision_tree_predicted])[0:2])
```

```python
# 연습문제 11 SVM, 결정 트리, 랜덤 포레스트 세 가지 기계 학습 모델 중에 어떤 것을 쓸지 결정하는 모델 선택을 해야한다고 가정하고 
# 5-겹 교차 검증을 사용하여 모델 선택을 수행한다.
# 모델을 선택하고 선택된 모델을 가지고 테스트 집합에 대한 성능을 측정하여 출력하는 프로그램을 작성하시오.
# 이때 sklearn이 제공하는 digit 데이터셋을 사용하시오.

from sklearn import datasets
from sklearn import svm 
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
import pandas as pd 
import numpy as np

digit = datasets.load_digits()

svm_model = svm.SVC(gamma = 0.1, C = 10)
svm_accuracies = cross_val_score(svm_model, digit.data, digit.target,cv=5)

decision_tree_model = DecisionTreeClassifier(random_state=1234)
decision_tree_accuracies = cross_val_score(decision_tree_model, digit.data, digit.target,cv=5)

random_forest_model = RandomForestClassifier(n_estimators=100,random_state=123456)
random_forest_accuracies = cross_val_score(random_forest_model, digit.data, digit.target,cv=5)

print(svm_accuracies)
print(decision_tree_accuracies)
print(random_forest_accuracies)
print("정확률(평균)=%0.3f%%, 표준편차=%0.3f"%(svm_accuracies.mean()*100,svm_accuracies.std()))
print("정확률(평균)=%0.3f%%, 표준편차=%0.3f"%(decision_tree_accuracies.mean()*100,decision_tree_accuracies.std()))
print("정확률(평균)=%0.3f%%, 표준편차=%0.3f"%(random_forest_accuracies.mean()*100,random_forest_accuracies.std()))
```

```python
# 번외 cross_val_score random 섞기
from sklearn import datasets
from sklearn import svm 
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold,StratifiedKFold
import pandas as pd 
import numpy as np

digit = datasets.load_digits()

kf = KFold(5, shuffle=True, random_state=1)
skf = StratifiedKFold(5, shuffle=True, random_state=1)

cv_lists = [5,kf,skf]

for cv in cv_lists:
  svm_model = svm.SVC(gamma = 0.1, C = 10)
  svm_accuracies = cross_val_score(svm_model, digit.data, digit.target,cv=cv)

  decision_tree_model = DecisionTreeClassifier(random_state=1234)
  decision_tree_accuracies = cross_val_score(decision_tree_model, digit.data, digit.target,cv=cv)

  random_forest_model = RandomForestClassifier(n_estimators=100,random_state=123456)
  random_forest_accuracies = cross_val_score(random_forest_model, digit.data, digit.target,cv=cv)

  print('===================================================================')
  print('cv = ', cv)
  print(svm_accuracies)
  print(decision_tree_accuracies)
  print(random_forest_accuracies)
  print("정확률(평균)=%0.3f%%, 표준편차=%0.3f"%(svm_accuracies.mean()*100,svm_accuracies.std()))
  print("정확률(평균)=%0.3f%%, 표준편차=%0.3f"%(decision_tree_accuracies.mean()*100,decision_tree_accuracies.std()))
  print("정확률(평균)=%0.3f%%, 표준편차=%0.3f"%(random_forest_accuracies.mean()*100,random_forest_accuracies.std()))
```
